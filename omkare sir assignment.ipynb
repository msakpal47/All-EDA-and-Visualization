```python
import math
import numpy as np


class stats:
    def __init__(self, data):
        self.data = data

    def sum_module(self):
        sum1 = sum(self.data)
        return sum1

    def len_module(self):
        len1 = len(self.data)
        return len1

    def mean_module(self):
        mean1 = np.mean(self.data)
        return mean1

    def median_module(self):
        median1 = np.median(self.data)
        return median1

    def mode_module(self):
        uniq_values = []
        mode_values = []
        frequency_count = {}

        for i in self.data:
            if i in frequency_count:
                frequency_count[i] += 1
                if i not in mode_values:
                    mode_values.append(i)
            else:
                frequency_count[i] = 1
                uniq_values.append(i)

        mode_count = {key: value for key, value in frequency_count.items() if value > 1}

        if not mode_count:
            return "No mode"
        else:
            return mode_count

    def variance_module(self):
        variance1 = np.var(self.data)
        return variance1

    def standard_deviation_module(self):
        variance1 = np.std(self.data)
        return variance1

    @staticmethod
    def covariance_module(list1, list2):
        if len(list1) != len(list2):
            raise ValueError("Lists must have the same length")

        mean1 = np.mean(list1)
        mean2 = np.mean(list2)

        cov = sum((x - mean1) * (y - mean2) for x, y in zip(list1, list2)) / len(list1)
        return cov

    @staticmethod
    def correlation_coefficient_module(data1, data2):
        correlation_matrix = np.corrcoef(data1, data2)[0, 1]
        return correlation_matrix

    def confidence_interval_z_module(self):
        mean = self.mean_module()
        std_dev = self.standard_deviation_module()
        n = len(self.data)

        z = 1.645
        margin_of_error = z * (std_dev / (math.sqrt(n)))
        lower_bound = mean - margin_of_error
        upper_bound = mean + margin_of_error

        z = 1.96
        margin_of_error2 = z * (std_dev / (math.sqrt(n)))

        lower_bound2 = mean - margin_of_error2
        upper_bound2 = mean + margin_of_error2

        z = 2.17
        margin_of_error3 = z * (std_dev / (math.sqrt(n)))

        lower_bound3 = mean - margin_of_error3
        upper_bound3 = mean + margin_of_error3

        return (f"\n 90% of Z-score is: {lower_bound}, {upper_bound}, "
                f"\n 95% of Z-score is: {lower_bound2}, {upper_bound2}, "
                f"\n 97% of Z-score is: {lower_bound3}, {upper_bound3}")

    def quartiles_module(self):
        q1 = np.percentile(self.data, 25)
        q2 = np.percentile(self.data, 50)
        q3 = np.percentile(self.data, 75)

        iqr = q3 - q1

        lower_quartile = q1 - 1.5 * iqr
        upper_quartile = q3 + 1.5 * iqr

        return q1, q2, q3, iqr, lower_quartile, upper_quartile


```
