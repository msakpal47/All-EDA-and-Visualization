```python
## removing the outlier 
```

```python
q1=df.quantile(0.0)   ## 25 % 
q3=df.quantile(0.90)  ## 75 %
iqr=q3-q1
```

```python
## to save in new dtaframe
df2=df[~((df<(q1-1.5*iqr))| (df>(q3+1.5*iqr)))] 
```

```python
df2.head(2)
```

```python
df3=df2.dropna()
```

```python
df3.shape
```

## removing the corelation matrix

```python
corr_matrix=df3.corr()
```

```python
columns = np.full((corr_matrix.shape[0],), True, dtype=bool)

## For loop to filter columns
for i in range(corr_matrix.shape[0]):
    for j in range(i+1, corr_matrix.shape[0]):
        if corr_matrix.iloc[i,j] >= 0.7:
            if columns[j]:
                columns[j] = False
                
# Selected column store into a variable
selected_columns = df3.columns[columns]
df_with_filtered_columns1 = df3[selected_columns]
```

```python
df_with_filtered_columns1.columns
```

```python
X11=df_with_filtered_columns1[['age', 'job', 'marital', 'education', 'default', 'housing', 'loan',
       'contact', 'month', 'day_of_week', 'duration', 'campaign', 'pdays',
       'previous', 'poutcome', 'emp.var.rate']]
```

```python
y11=df_with_filtered_columns1["y"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X11,y11,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.predict(X_test)
```

```python
y_predict=model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
clf=classification_report(y_test,y_predict)
print(clf)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_predict)
cm
```

```python
sn.heatmap(cm,annot=True)
```

```python
## roc curve is alos know as auc (tp vs fp) / precison recall cure (tp vs predicted positive) 
```

```python
##   GridSearchCV
```

GridSearchCV shows how a classifier is optimized by cross-validation.

When the grid search is called with various parameters, it chooses the one with the highest score based on the given Random Forest model. 'bestparams' gives the information of the parameters that resulted in the highest score.

```python
param_grid = {
    'max_depth': [2,4,8,10],
    'min_samples_leaf': range(1,2,3),
    'min_samples_split': range(2,4,6)}
```

```python
from sklearn.model_selection import GridSearchCV
grid= GridSearchCV(model, param_grid, cv= 5)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.score(X_test,y_test)
```

```python
grid.best_estimator_
```

```python
grid.best_params_
```

```python
grid.best_score_
```
